```julia
function my_plot(punkt)
    my_max = max(abs.(punkt))
    if punkt[1] < 0
        my_xlims = (-my_max, 0)
    else
        my_xlims = (my_max, 0)
    end
     if punkt[2] < 0
        my_ylims = (-my_max, 0)
    else
        my_ylims = (my_max, 0)
    end
    plot([0,punkt[1]], [0,punkt[2]] , arrow = true, xlims= my_xlims, ylims = my_ylims)
end
```

```julia
my_plot([1;1])
```

```julia
A = [1 0; 0 1]
```

```julia
B = [0 1; 1 0]
```

```julia
my_plot(A*[1;1])
```

```julia
my_plot([-1;1])
```

```julia
my_plot(A*[])
```
