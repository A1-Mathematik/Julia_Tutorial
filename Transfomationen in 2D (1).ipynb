```julia
function my_plot(punkt)
    my_max = maximum(abs.(punkt))
    if punkt[1] < 0
        my_xlims = (-8.4, 0)./(3/my_max)
    else
        my_xlims = (0, 8.4)./(3/my_max)
    end
     if punkt[2] < 0
        my_ylims = (-5.6, 0)./(3/my_max)
    else
        my_ylims = (0, 5.6)./(3/my_max)
    end
    plot([0,punkt[1]], [0,punkt[2]] , arrow = true, xlims= my_xlims, ylims = my_ylims)
    end
```

```julia
my_plot([10;10])
```
